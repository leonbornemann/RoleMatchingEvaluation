```python
import pandas as pd
from plotnine import *
from plotnine.data import *
from tabulate import tabulate
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix, precision_recall_curve
import numpy as np

df = pd.read_csv("localData/blockingResultSize/simpleBlockingResultSize.csv")
dfFinalRows = []

for dsName in df['dataset']:
    for col in df.columns:
        if(col != "dataset"):
            count = df[df['dataset']==dsName][col].iloc[0]
            dfFinalRows.append({'dataset' : dsName, 'method' : col , "count" : count})

dfFinal = pd.DataFrame(dfFinalRows)

# dfFinal['method_ordered']=totalAggregateDF['dataset']
# totalAggregateDF['dataset_unordered']= list(map(lambda x : datasetToAbbreviation[x],list(totalAggregateDF['dataset_long'])))
#
#
#
# dfFinal['method_ordered']= list(map(lambda x : datasetToAbbreviation[x],list(totalAggregateDF['dataset_long'])))
# dfFinal['method_ordered'] = pd.Categorical(dfFinal['method'], ["EM","CSM","VSM","OR","UT","ED","FO","MI","PO","TV"])
#
# totalAggregateDF['dataset'] = pd.Categorical(totalAggregateDF['dataset_unordered'], ["AU","CH","MD","OR","UT","ED","FO","MI","PO","TV"])



plot = ggplot(dfFinal,aes(x = 'method',y='count',fill='dataset')) + geom_col(position=position_dodge()) + scale_y_log10()
print(plot)

plot = ggplot(dfFinal,aes(x = 'dataset',y='count',fill='method')) + geom_col(position=position_dodge()) + scale_y_log10()
print(plot)

plot = ggplot(dfFinal,aes(x = 'method',y='count',color='dataset')) + geom_point() + geom_line() + scale_y_log10()
print(plot)
```

# By Dataset

```python
for ds in dfFinal['dataset'].unique():
    dfForThisDataset = dfFinal[dfFinal['dataset']==ds]
    plot = ggplot(dfForThisDataset,aes(x = 'method',y='count',fill='method')) + ggtitle(ds) + geom_col(position=position_dodge()) + scale_y_log10()
    print(plot)
```

# BY Method

```python
for method in dfFinal['method'].unique():
    dfForThisDataset = dfFinal[dfFinal['method']==method]
    plot = ggplot(dfForThisDataset,aes(x = 'dataset',y='count',fill='dataset')) + ggtitle(method) + geom_col(position=position_dodge()) + scale_y_log10()
    print(plot)
```
